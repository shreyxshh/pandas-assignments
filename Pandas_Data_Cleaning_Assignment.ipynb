## Q1. Find Missing Values

Input:
df = pd.DataFrame({

## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
# Write your code here
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}
df = pd.DataFrame(data)

#filling age with mean value
df['Age'] = df["Age"].fillna(df.Age.mean())

#filling null name with unknown
df['Name'] = df["Name"].fillna('Unknown')

#removing duplicate rows
df.drop_duplicates(inplace=True, keep='first')

#final dataframe
df
```
