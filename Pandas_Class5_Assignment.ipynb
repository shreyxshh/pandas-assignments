## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
# Write your code here
import pandas as pd
df = pd.DataFrame({
    'Age' : [25, 18, 30, 22]
})
#ascending order
(df.sort_values(by='Age'))
```

## Q8. Count City Frequency

Input:
City = ['Bhopal','Indore','Bhopal','Jabalpur']

Output:
Bhopal      2
Indore      1
Jabalpur    1

```python
# Write your code here
df = pd.DataFrame({
    'City' : ['Bhopal', 'Indore','Bhopal', 'Jabalpur']
})

#to count the freq of elements in any df
print(df['City'].value_counts())
```

## Q9. Find Percentage Frequency

Input:
City = ['A','A','B','C']

Output:
A = 50%
B = 25%
C = 25%

```python
# Write your code here
df = pd.DataFrame({
    'City' : ['A','A','B','C']
})

#gives the percentage frequency of element in df
print(df['City'].value_counts(normalize=True)*100)
```

## Q11. Find Correlation

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
1.0

```python
# Write your code here
df1 = pd.DataFrame({
    'Math' : [10,20,30,40],
    'Science' : [15,25,35,45]
})
#normal correlation 
print(df1.corr(numeric_only=True))
#relative correlation
print(df1.Math.corr(df1.Science))
```

## Q12. Find Covariance

Input:
Math = [10,20,30,40]
Science = [15,25,35,45]

Output:
166.67 (approx)

```python
# Write your code here
#normal covariance
print(df1.cov(numeric_only=True))
#relative convariance 
print(df1.Math.cov(df1.Science))
```

## Q14. Student Statistics

Input:
Marks = [60,70,80,90]

Output:
Mean = 75
Median = 75
Max = 90
Min = 60

```python
# Write your code here
student = pd.DataFrame({
    'marks' : [60,70,80,90]
})
print(student.marks.mean())
print(student.marks.median())
print(student.marks.max())
print(student.marks.min())
```
