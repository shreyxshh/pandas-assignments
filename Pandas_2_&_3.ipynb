
# Q8. Handle Missing Values

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman", None, "Neha"]
Marks = [80, 75, None]
```

Fill missing values using `fillna()`.

## Sample Output
```python
Missing values replaced
```


````python
# Write your code here
df = pd.DataFrame({
    'Name' : ['Aman', None, 'Neha'],
    'Marks' : [80, 75, None]
})

#avoiding the use of inplace
df['Name'] = df["Name"].fillna('Unknown')
df['Marks'] = df['Marks'].fillna(0)
df
````


# Q9. Replace Values

## Problem Statement
Create a DataFrame:

```python
City = ["BPL", "IND", "BPL"]
```

Replace:
- BPL → Bhopal
- IND → Indore

## Sample Output
```python
Bhopal
Indore
Bhopal
```


````python
# Write your code here
df = pd.DataFrame({
    'City' : ["BPL", "IND", "BPL"]
})
df['City'] = df['City'].replace(['BPL', 'IND'], ['Bhopal', 'Indore'])
print(df)
````


# Q10. Remove Duplicate Records

## Problem Statement
Create a DataFrame:

```python
Name = ["Aman","Rahul","Aman","Neha"]
```

Check duplicate rows using `duplicated()` and remove them using `drop_duplicates()`.

## Sample Output
```python
Aman
Rahul
Neha
```


````python
# Write your code here
df = pd.DataFrame({
    "Name" : ['Aman', 'Rahul', 'Aman' ,"Neha"]
})
print(df.duplicated())
df.drop_duplicates(inplace=True, keep='first')
print(df)
````
